```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.datasets import load_digits
digits=load_digits()

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.DataFrame(digits.data)
df["target"]=digits.target
df["target_name"]=df.target.apply(lambda x:digits.target_names[x])
df.head()
```

```python
model_para=({
    "LogisticRegression": {
        "model": LogisticRegression(solver="liblinear", multi_class="auto"),
        "para": {
            "C": [5,10,20]  
          } 
        },
        "RandomForestClassifier": {
            "model": RandomForestClassifier(),
            "para": {
                "n_estimators": [5,10,20],
                "criterion": ["gini","entropy","log_loss"]
              }
            },
            "SVC": {
                "model": SVC(gamma="auto"),
                "para": {
                    "C": [1,10,20],
                    "kernel": ["linear","poly","rbf","sigmoid"]
                  }
                },
                "MultinomialNB": {
                    "model": MultinomialNB(),
                    "para": {
                        "alpha": [0,1]
                      }
                    },
                    "GaussianNB": {
                        "model": GaussianNB(),
                        "para": { }
                    },
                        "DecisionTreeClassifier": {
                            "model": DecisionTreeClassifier(),
                            "para": {
                                "criterion": ["gini","entropy","log_loss"]
                            }
                        }         
})
```

```python
scores= []

from sklearn.model_selection import GridSearchCV
for model_name, mp in model_para.items():
  gscv=GridSearchCV(mp["model"], mp["para"], cv=5, return_train_score=False)
  gscv.fit(digits.data, digits.target)
  scores.append({
      "model name": model_name,
      "best score": gscv.best_score_,
      "best parameter": gscv.best_params_
  })

df=pd.DataFrame(scores, columns=["model name","best score","best parameter"])
df
```

```python
from sklearn.datasets import load_wine
wine=load_wine()
```

```python
df=pd.DataFrame(wine.data)
df["target"]=wine.target
df["target_name"]=df.target.apply(lambda x:wine.target_names[x])
df.head()
```

```python
model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    },
    'naive_bayes_gaussian': {
        'model': GaussianNB(),
        'params': {}
    },
    'naive_bayes_multinomial': {
        'model': MultinomialNB(),
        'params': {}
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini','entropy'],
            
        }
    }     
}
```

```python
from sklearn.model_selection import GridSearchCV
import pandas as pd
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(wine.data, wine.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
