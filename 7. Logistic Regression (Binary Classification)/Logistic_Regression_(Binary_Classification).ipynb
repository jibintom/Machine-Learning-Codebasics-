```python
iv=df1[["satisfaction_level","average_montly_hours","promotion_last_5years"]]
iv
```

```python
salary_con=pd.get_dummies(df1.salary,prefix="salary")
salary_con
```

```python
 x=pd.concat([iv,salary_con],axis=1)
 x
```

```python
y=df1.left
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
y_predict=lg.predict(x_test)
y_predict
```

```python
lg.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print("r2_score is",r2_score(y_test,y_predict))
print("mean_squared_error is",mean_squared_error(y_test,y_predict,squared=False))
print("mean_squared_error is",mean_squared_error(y_test,y_predict))
```
